```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
df = pd.read_csv("../data/insurance.csv")
df.head()
```

```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns

num_cols, cat_cols
```

```python
df[num_cols].hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
for col in cat_cols:
    sns.countplot(x=df[col])
    plt.title(col)
    plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["charges"], kde=True)
plt.title("Distribution of Medical Insurance Charges")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="smoker", y="charges", data=df)
plt.title("Charges vs Smoking")
plt.show()
```

```python
sns.scatterplot(x="age", y="charges", data=df)
plt.show()
```

```python
sns.scatterplot(x="bmi", y="charges", data=df)
plt.show()
```

### EDA Insights

1. Smoking has a strong impact on medical insurance charges.
2. Charges increase with age, especially for smokers.
3. BMI shows a moderate correlation with charges.
4. The target variable is right-skewed.
